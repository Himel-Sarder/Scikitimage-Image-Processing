```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

edges = cv2.Canny(image, 100, 200)
# cv2.Canny(image, threshold1, threshold2)
# threshold1 (Low threshold)
#       Detects weak edges
# threshold2 (High threshold)
#       Detects strong edges

plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```

```python
image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

edges = cv2.Canny(image, 20, 50)
# cv2.Canny(image, threshold1, threshold2)
# threshold1 (Low threshold)
#       Detects weak edges
# threshold2 (High threshold)
#       Detects strong edges

plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```

```python
edges1 = cv2.Canny(image, 5, 15)
edges2 = cv2.Canny(image, 10, 20)
edges3 = cv2.Canny(image, 25, 50)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(edges1, cmap='gray')
plt.title('5-15')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(edges2, cmap='gray')
plt.title('10-20')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(edges3, cmap='gray')
plt.title('25-50')
plt.axis('off')

plt.show()
```
