# Logarithmic Transformation

# Logarithmic Transformation is a powerful point processing technique used to enhance images—especially when pixel intensity values vary widely.
# The log transformation maps a narrow range of low input grey level values into a wider range of output values
# The inverse log transformation performs the opposite transformation 
# Log functions are particularly useful when the input grey level values may have an extremely large range of values

# Why Log Transformation is Used 

# Behavior of Log Curve 


# Real-Life Applications 



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image (grayscale)
img = cv2.imread('blur.jpg', 0)

# Convert to float
img_float = np.float32(img)

# Apply log transformation
c = 255 / np.log(1 + np.max(img_float))
log_img = c * np.log(1 + img_float)

# Convert back to uint8
log_img = np.uint8(log_img)

# Show results
plt.imshow(log_img, cmap='gray')
plt.title("Log Transformed Image")
plt.axis('off')
plt.show()
```
