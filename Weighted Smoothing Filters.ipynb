# Weighted smoothing filters are an extension of the basic mean (averaging) filter. Instead of giving equal importance to all neighbouring pixels, they assign different weights to different pixels in the neighbourhood.

## This allows better smoothing while preserving important details like edges more effectively than simple averaging.

## In a weighted smoothing filter:
### - Each pixel in the kernel has a specific weight
### - The center pixel usually has higher weight
### - Nearby pixels contribute more than distant ones
## The output pixel is computed as a weighted average.


## Example Kernel (3×3 Weighted Filter) 

## Gaussian Filter (Most Important Weighted Filter)
## The most common weighted smoothing filter is based on the concept of Gaussian distribution.
### Properties:
### - Weights follow a bell-shaped curve
### - Closer pixels have higher influence
### - Produces natural-looking blur

## Advantages Over Mean Filter
### - Better edge preservation
### - Less blurring of important details
### - More realistic smoothing
### - Reduces noise more effectively in many cases
## Disadvantages
### - Slightly more computationally expensive
### - Requires proper selection of weights

```python
import cv2
import matplotlib.pyplot as plt

# Load image in grayscale
image = cv2.imread('image2.jpg', cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Error: Image not found!")
    exit()

# Apply Gaussian weighted smoothing
gaussian = cv2.GaussianBlur(image, (3, 3), 0)

# Plot
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Grayscale Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gaussian, cmap='gray')
plt.title('Gaussian Smoothed Image (3x3)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Load image
# -------------------------------
image = cv2.imread('image2.jpg', cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Error: Image not found!")
    exit()

# Convert BGR to RGB
# image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# -------------------------------
# 1. Gaussian Weighted Filter
# -------------------------------
gaussian = cv2.GaussianBlur(image, (5, 5), 0)
# gaussian_rgb = cv2.cvtColor(gaussian, cv2.COLOR_BGR2RGB)

# -------------------------------
# 2. Custom Weighted Kernel
# -------------------------------
kernel = np.array([[1, 2, 1],
                   [2, 4, 2],
                   [1, 2, 1]], dtype=np.float32) / 16

weighted = cv2.filter2D(image, -1, kernel)
# weighted_rgb = cv2.cvtColor(weighted, cv2.COLOR_BGR2RGB)

# -------------------------------
# Plot results
# -------------------------------
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(gaussian, cmap='gray')
plt.title('Gaussian Weighted Filter (5x5)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(weighted, cmap='gray')
plt.title('Custom Weighted Filter (3x3)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
