```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

blur = cv2.GaussianBlur(image, (5, 5), 0)
# cv2.GaussianBlur(src, ksize, sigmaX)
# ksize (Kernel size)
#    Must be odd: (3,3), (5,5), (7,7)
#    Larger kernel → more blur
# sigmaX (Standard deviation)
#    Controls spread of Gaussian
#    If 0 → OpenCV calculates automatically

plt.imshow(blur, cmap='gray')
plt.title('Gaussian Blurred Image')
plt.axis('off')
plt.show()
```

```python
blur_3 = cv2.GaussianBlur(image, (3,3), 0)
blur_7 = cv2.GaussianBlur(image, (7,7), 0)
blur_15 = cv2.GaussianBlur(image, (15,15), 0)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(blur_3, cmap='gray')
plt.title('3x3')

plt.subplot(1,3,2)
plt.imshow(blur_7, cmap='gray')
plt.title('7x7')

plt.subplot(1,3,3)
plt.imshow(blur_15, cmap='gray')
plt.title('15x15')

plt.show()
```

```python
noiseimage = cv2.imread('noise_image.jpg', cv2.IMREAD_GRAYSCALE)

blur = cv2.GaussianBlur(noiseimage, (5, 5), 0)

plt.imshow(noiseimage, cmap='gray')
plt.title('Original')
```

```python
plt.imshow(blur, cmap='gray')
plt.title('Gaussian Blurred Image')
plt.show()
```

```python
import numpy as np

def add_gaussian_noise(img, sigma=25):
    noise = np.random.normal(0, sigma, img.shape)
    noisy = img + noise
    return np.clip(noisy, 0, 255).astype('uint8')

noisy = add_gaussian_noise(noiseimage)

denoised = cv2.GaussianBlur(noisy, (5,5), 0)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(noiseimage, cmap='gray')
plt.title('Original')

plt.subplot(1,3,2)
plt.imshow(noisy, cmap='gray')
plt.title('Noisy')

plt.subplot(1,3,3)
plt.imshow(denoised, cmap='gray')
plt.title('Denoised')

plt.show()
```
