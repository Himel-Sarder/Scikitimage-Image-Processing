# The Laplacian filter is a second-order derivative spatial filter used mainly for edge detection and image sharpening. It highlights regions of rapid intensity change (edges) in an image.
## - It measures how much a pixel differs from its neighbours.
## - Unlike first-order derivatives (like Sobel), the Laplacian detects edges in all directions.
## - It is isotropic (direction-independent).






## Simplified Image Enhancement



## Applications 
### - Edge detection
### - Image sharpening
### - Feature extraction
### - Medical image analysis

```python
import cv2
import matplotlib.pyplot as plt

# -------------------------------
# Load grayscale image
# -------------------------------
image = cv2.imread('image2.jpg', cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Error: Image not found!")
    exit()

# -------------------------------
# Apply Gaussian blur (optional, reduces noise)
# -------------------------------
blur = cv2.GaussianBlur(image, (3, 3), 0)

# -------------------------------
# Apply Laplacian filter
# -------------------------------
laplacian = cv2.Laplacian(blur, cv2.CV_64F)

# Convert to absolute scale for display
laplacian_abs = cv2.convertScaleAbs(laplacian)

# -------------------------------
# Create sharpened image
# sharpened = original - laplacian
# -------------------------------
sharpened = cv2.subtract(image, laplacian_abs)

# -------------------------------
# Plot results
# -------------------------------
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(laplacian_abs, cmap='gray')
plt.title('Laplacian Edge')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sharpened, cmap='gray')
plt.title('Sharpened Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
# Load grayscale image
# -------------------------------
image = cv2.imread('image2.jpg', cv2.IMREAD_GRAYSCALE)

if image is None:
    print("Error: Image not found!")
    exit()

# -------------------------------
# Apply Gaussian blur (optional, reduces noise)
# -------------------------------
blur = cv2.GaussianBlur(image, (3, 3), 0)

# -------------------------------
# Define Laplacian kernel (8-neighbor)
# -------------------------------
laplacian_kernel = np.array([[ -1, -1, -1],
                             [ -1,  8, -1],
                             [ -1, -1, -1]], dtype=np.float32)

# -------------------------------
# Apply Laplacian filter using kernel
# -------------------------------
laplacian = cv2.filter2D(blur, cv2.CV_64F, laplacian_kernel)

# Convert to absolute scale for display
laplacian_abs = cv2.convertScaleAbs(laplacian)

# -------------------------------
# Create a sharpened image
# sharpened = original - laplacian
# -------------------------------
sharpened = cv2.subtract(image, laplacian_abs)

# -------------------------------
# Plot results
# -------------------------------
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(laplacian_abs, cmap='gray')
plt.title('Laplacian Edge (Kernel)')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(sharpened, cmap='gray')
plt.title('Sharpened Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```
