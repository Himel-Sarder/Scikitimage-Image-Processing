```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('bonefructure.jpg', 0)

# Normalize image (0 to 1)
img_norm = img / 255.0

gamma = 0.2 

power_img = np.power(img_norm, gamma)

# Scale back to 0-255
power_img = np.uint8(power_img * 255)

# Plot both images
plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.axis('off')

# Gamma Transformed Image
plt.subplot(1, 2, 2)
plt.imshow(power_img, cmap='gray')
plt.title(f"Gamma = {gamma}")
plt.axis('off')

plt.show()
```

```python
gamma = 0.5

power_img = np.power(img_norm, gamma)

# Scale back to 0-255
power_img = np.uint8(power_img * 255)

plt.imshow(power_img, cmap='gray')
plt.title(f"Gamma = {gamma}")
plt.axis('off')
plt.show()
```
