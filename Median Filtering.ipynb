```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('image.jpg', cv2.IMREAD_GRAYSCALE)

median = cv2.medianBlur(image, 5)
# cv2.medianBlur(src, ksize)
# ksize (Kernel size)
    # Must be odd: 3, 5, 7, 9
    # Larger kernel → stronger noise removal
    # Too large → loss of fine details

plt.imshow(median, cmap='gray')
plt.title('Median Filtered Image')
plt.axis('off')
plt.show()
```

```python
median_3 = cv2.medianBlur(image, 3)
median_7 = cv2.medianBlur(image, 7)
median_15 = cv2.medianBlur(image, 15)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(median_3, cmap='gray')
plt.title('3x3')

plt.subplot(1,3,2)
plt.imshow(median_7, cmap='gray')
plt.title('7x7')

plt.subplot(1,3,3)
plt.imshow(median_15, cmap='gray')
plt.title('15x15')

plt.show()
```

```python
import numpy as np

def add_salt_pepper(img, prob=0.02):
    noisy = img.copy()
    h, w = img.shape

    num_pixels = int(prob * h * w)

    for _ in range(num_pixels // 2):
        x = np.random.randint(0, h)
        y = np.random.randint(0, w)
        noisy[x, y] = 255

    for _ in range(num_pixels // 2):
        x = np.random.randint(0, h)
        y = np.random.randint(0, w)
        noisy[x, y] = 0

    return noisy

noisy = add_salt_pepper(image)

denoised = cv2.medianBlur(noisy, 5)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.imshow(image, cmap='gray')
plt.title('Original')

plt.subplot(1,3,2)
plt.imshow(noisy, cmap='gray')
plt.title('Noisy')

plt.subplot(1,3,3)
plt.imshow(denoised, cmap='gray')
plt.title('Denoised')

plt.show()
```

## Build-in

```python
image = cv2.imread('image.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply Median Filter
median = cv2.medianBlur(image, 5)

# Plot
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(median)
plt.title('Median Filter')
plt.axis('off')

plt.show()
```
