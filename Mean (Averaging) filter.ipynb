# The mean filter (also called the averaging filter) is one of the simplest linear spatial filters used in image processing. It is mainly applied for smoothing and noise reduction.
## The mean filter works by replacing each pixel value with the average (mean) of its neighboring pixel values, including itself.
## This process reduces intensity variations between adjacent pixels, which results in a smoother image.

## Applications


# Using cv2.blur() (Built-in Mean Filter)

```python
import cv2
import matplotlib.pyplot as plt

# Load image
image = cv2.imread('image2.jpg')

# Convert BGR (OpenCV) to RGB (Matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Apply mean filter (3x3 kernel)
mean_filtered = cv2.blur(image, (3, 3))

# Convert filtered image to RGB
mean_filtered_rgb = cv2.cvtColor(mean_filtered, cv2.COLOR_BGR2RGB)

# Plot images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(mean_filtered_rgb)
plt.title('Mean Filtered Image (3x3)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

# Manual

## 3x3 mean kernel

```python
import numpy as np
# Convert BGR to RGB for correct display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Create 3x3 mean kernel
kernel = np.ones((3, 3), np.float32) / 9

# Apply filter
mean_filtered = cv2.filter2D(image, -1, kernel)

# Convert filtered image to RGB
mean_filtered_rgb = cv2.cvtColor(mean_filtered, cv2.COLOR_BGR2RGB)

# Plot using matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(mean_filtered_rgb)
plt.title('Mean Filtered Image (3x3 Kernel)')
plt.axis('off')

plt.tight_layout()
plt.show()
```
