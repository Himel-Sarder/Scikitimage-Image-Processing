# Binary thresholding is a fundamental image processing technique that converts grayscale images into binary (black and white) images by comparing pixel intensities against a set threshold value.


# Mathematical Definition 

# Conceptual Interpretation 
## Binary thresholding creates a decision boundary in intensity space:
## - Pixels with intensity below 𝑇 are mapped to background (0)
## - Pixels with intensity above or equal to 𝑇 are mapped to foreground (255)
## This effectively segments the image into two classes.

# Histogram-Based Understanding
## Thresholding is most effective when the grayscale histogram is bimodal:
## - One peak represents background
## - One peak represents foreground
## The threshold is ideally chosen at the valley between the peaks, maximizing class separation.
# Types of Binary Thresholding 


```python
import cv2
import matplotlib.pyplot as plt

# Read image
img = cv2.imread('blur.jpg')

# Convert to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply threshold
ret, thresh = cv2.threshold(gray, 100, 125, cv2.THRESH_BINARY)

# Plot using matplotlib
plt.figure(figsize=(10,5))

# Original Image
plt.subplot(1,2,1)
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

# Binary Image
plt.subplot(1,2,2)
plt.imshow(thresh, cmap='gray')
plt.title("Binary Threshold")
plt.axis('off')

plt.show()
```

# ret, output = cv2.threshold(src, T, maxVal, type)
### - src: input grayscale image
### - T: threshold value
### - maxVal: value assigned to pixels above threshold
### - type: thresholding method
### - ret: returned threshold (useful in automatic methods)

# Threshold Selection Methods 
## Manual Threshold
### -> Fixed value (e.g., 127)
### -> Works only if lighting and contrast are stable
## Otsu’s Method
## -> Automatically computes optimal threshold by minimizing intra-class variance
## -> Suitable for bimodal histograms
 

## Adaptive Thresholding
### -> Computes threshold locally for each region
### -> Effective under non-uniform illumination
